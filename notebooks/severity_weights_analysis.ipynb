## Generating the Weights


```python
import json
from pathlib import Path

from scripts.assessment_data.compute_severity_weights import compute_weights

repo_root = Path.cwd().parent
data_path = repo_root / 'data' / 'clean_ipp_data.csv'
out_path = repo_root / 'config' / 'weights' / 'severity_weights.json'

weights = compute_weights(data_path)

out_path.parent.mkdir(parents=True, exist_ok=True)
with out_path.open('w', encoding='utf-8') as fh:
    json.dump(weights, fh, indent=2, ensure_ascii=True)
    fh.write("\n")
```
